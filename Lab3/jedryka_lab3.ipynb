# Zadanie 1

```python
def find_cost(s1,s2):
    n , m = len(s1) + 1 , len(s2) + 1
    board = [[0 for i in range(m)] for j in range(n)]
    
    for i in range(m):
        board[0][i] = i
    for i in range(1,n):
        board[i][0] = i

    for i in range(1,n):
        for j in range(1,m):
            if s1[i-1] == s2[j-1]:
                cost = 0
            else:
                cost = 1
            board[i][j] = min(
                board[i-1][j] + 1,
                board[i][j-1] + 1,
                board[i-1][j-1] + cost
            )
    
    return board

def Levenshtein(s1,s2):
    board = find_cost(s1,s2)
    changes = []
    a = len(s1)
    b = len(s2)

    while a != 0 or b!=0:
        if board[a-1][b] + 1 == board[a][b]:
            changes.append("delete")
            a-=1
        elif board[a][b-1] + 1 == board[a][b]:
            changes.append("add")
            b-=1
        elif board[a-1][b-1] == board[a][b]:
            changes.append("leave")
            a-=1
            b-=1
        else:
             changes.append("swap")
             a-=1
             b-=1
    
    return board , changes
```

# Zadanie 2

```python
def change_word(s1,s2):
    board , changes = Levenshtein(s1 ,s2)
    x,y = -1,-1
    current = s1
    print(f"Changing {s1} to {s2}:")

    for action in reversed(changes):
        x+=1
        y+=1
        if action == "leave":
            continue
        elif action == "delete":
            if x < len(current)-1:
                print(f"Delete {current[x]}:  {current} -> {current[:x]+current[x+1:]}")
                current = current[:x]+current[x+1:]
            else:
                print(f"Delete {current[x]}:  {current} -> {current[:x]}")
                current = current[:x]
        elif action == "add":
             print(f"Add {s2[y]}: {current} -> {current[:y]+s2[y]+current[y:]}")
             current = current[:y]+s2[y]+current[y:]
        else:
            print(f"Replace {current[x]} -> {s2[y]}:  {current} -> {current[:x]+s2[y]+current[x+1:]}")
            current = current[:x]+s2[y]+current[x+1:]
    
    print(f"Total editing distance is {board[len(s1)][len(s2)]}")
```

```python
def LCS(s1,s2):
    n , m = len(s1) + 1 , len(s2) + 1
    board = [[0 for i in range(m)] for j in range(n)]
    
    for i in range(m):
        board[0][i] = 0
    for i in range(1,n):
        board[i][0] = 0

    for i in range(1,n):
        for j in range(1,m):
            if s1[i-1] == s2[j-1]:
                board[i][j] = board[i-1][j-1]+1
            else:
                board[i][j] = max(board[i-1][j],board[i][j-1])
    
    return board[n-1][m-1]
```

```python
import random
def delete_random_tokens(tokens):
    output = []

    for token in tokens:
        roll = random.uniform(0.0,1.0)
        if roll > 0.03:
            output.append(token)
    return output
```

```python
txt1 = delete_random_tokens(tokens)
txt2 = delete_random_tokens(tokens)
```
